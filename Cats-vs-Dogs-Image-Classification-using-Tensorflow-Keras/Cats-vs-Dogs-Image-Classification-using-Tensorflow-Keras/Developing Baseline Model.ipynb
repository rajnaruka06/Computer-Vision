#### We'll create a baseline model quickly and then iterate through it to get better results.

```python
## Load dataset that we'd created.
import pickle
X = pickle.load(open("X.pickle", "rb"))
y = pickle.load(open("y.pickle", "rb"))
```

- Normalization(x) = (x - x.min)/(x.max - x.min) here x.max == 255, x.min == 0 
- Normalization(x) = x/255.0

```python
import numpy as np
X = X/255.0
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2)

print(X_train.shape, X_test.shape)
## ip_shape = (X.shape[1],X.shape[2], 1)
```

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential()
ip_shape = (X.shape[1], X.shape[2], 1)

model.add(tf.keras.layers.Conv2D(8, kernel_size=3, input_shape=X.shape[1:]))
model.add(tf.keras.layers.MaxPool2D((2, 2)))

model.add(tf.keras.layers.Conv2D(16, kernel_size=3))
model.add(tf.keras.layers.MaxPool2D((2, 2), strides=2))

model.add(tf.keras.layers.Conv2D(32, kernel_size=3))
model.add(tf.keras.layers.MaxPool2D((2, 2), strides=2))


model.add(tf.keras.layers.Flatten())


model.add(tf.keras.layers.Dense(64, 'relu', kernel_regularizer = "L2"))
model.add(tf.keras.layers.Dense(32, 'relu', kernel_regularizer = "L2"))
model.add(tf.keras.layers.Dense(16, 'relu', kernel_regularizer = "L2"))
model.add(tf.keras.layers.Dense(1, 'sigmoid'))

model.compile('adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=200, use_multiprocessing=True)
```

```python
from sklearn.metrics import f1_score, accuracy_score
```

```python
pred = model.predict(X_train)
pred1 = np.apply_along_axis(lambda x : int(x>0.5), 1, pred)
pred1 = pred1.reshape(pred1.shape[0], 1)
print(f1_score(pred1, y_train), accuracy_score(pred1, y_train))
```

```python
pred = model.predict(X_test)
pred1 = np.apply_along_axis(lambda x : int(x>0.5), 1, pred)
pred1 = pred1.reshape(pred1.shape[0], 1)
print(f1_score(pred1, y_test), accuracy_score(pred1, y_test))
```

### Second Model
- Let's try adding Reggularization.

```python
model = tf.keras.models.Sequential()
ip_shape = (X.shape[1], X.shape[2], 1)

model.add(tf.keras.layers.Conv2D(8, kernel_size=3, input_shape=X.shape[1:]))
model.add(tf.keras.layers.MaxPool2D((2, 2)))

model.add(tf.keras.layers.Conv2D(16, kernel_size=3))
model.add(tf.keras.layers.MaxPool2D((2, 2), strides=2))

model.add(tf.keras.layers.Conv2D(32, kernel_size=3))
model.add(tf.keras.layers.MaxPool2D((2, 2), strides=2))


model.add(tf.keras.layers.Flatten())


model.add(tf.keras.layers.Dense(64, 'relu', kernel_regularizer = "L2"))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(32, 'tanh', kernel_regularizer = "L2"))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(16, 'relu', kernel_regularizer = "L2"))
model.add(tf.keras.layers.Dense(1, 'sigmoid'))

model.compile('adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs = 200, use_multiprocessing = True)
```

```python
pred = model.predict(X_test)
pred1 = np.apply_along_axis(lambda x : int(x>0.5), 1, pred)
pred1 = pred1.reshape(pred1.shape[0], 1)
print(f1_score(pred1, y_test), accuracy_score(pred1, y_test))
```
